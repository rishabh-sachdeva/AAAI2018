<h1>************Features into log files*************</h1>

```python
import os
for key in gld_vis:
    arr=gld_vis[key][0].detach().numpy()
    #print(arr)
    category = key.rsplit('_',2)[0]
    instance = key.rsplit('_',1)[0]
    folder = 'visual_features/'+category+'/'+instance
    if not os.path.exists(folder):
        os.makedirs(folder)
    with open(folder+'/'+instance+'_object.log','a') as file:
        content = (",".join(map(str, arr)))
        file.write(content+'\n')
```
