```python
df_audio.AUDIO_FILE=df_audio.AUDIO_FILE.astype(str)
```

```python
df_audio["AUDIO_FILE"]= df_audio["AUDIO_FILE"].str.split("-",1, expand = True) 
```

```python
df_audio.head(10)
```

```python
df_audio["object"]=df_audio["AUDIO_FILE"].str.split('_',n=1,expand=True)[0]
```

```python
df_audio['instance'] = df_audio.object + '/'+df_audio.AUDIO_FILE
```

```python
df_audio.drop('AUDIO_FILE',axis=1,inplace=True)
df_audio.drop('object',axis=1,inplace=True)
```

```python
df_audio = df_audio[['instance','RESPONSE']]
```

```python
df_audio.to_csv('audio_descriptions.csv')
```
